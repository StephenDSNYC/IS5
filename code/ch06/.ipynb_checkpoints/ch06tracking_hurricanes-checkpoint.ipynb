```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
mpl.rcParams['font.family'] = 'serif'  # for matplotlib font configuration
```

```python
datafile = '../../dataset/ch06/tracking_hurricanes_2015.txt'
hurricanes = pd.read_csv(datafile,
                         sep='\t')
```

```python
hurricanes.info()
```

```python
# how to plot scatterplot
plt.figure(figsize=(8,3))
plt.scatter(hurricanes.Year, hurricanes.Error_72h,
            s=8, marker='o');  # marker is type of marker, s is marker Size
plt.xlabel('Year')
plt.ylabel('72h Error')
```

```python
# scatterplot and LOWESS
plt.figure(figsize=(8,3))
plt.scatter(hurricanes.Year, hurricanes.Error_72h,
            s=8, marker='o');
# note that y is lowess-ed against x, and y comes first
smoothed_error_72h = sm.nonparametric.lowess(exog=hurricanes.Year, endog=hurricanes.Error_72h)
plt.plot(smoothed_error_72h[:, 0], smoothed_error_72h[:, 1], 'r');
```
