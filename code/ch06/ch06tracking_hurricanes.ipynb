```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
mpl.rcParams['font.family'] = 'serif'
```

```python
datafile = '../../dataset/ch06/tracking_hurricanes_2015.txt'
hurricanes = pd.read_csv(datafile,
                         sep='\t')
```

```python
hurricanes.info()
```

```python
plt.figure(figsize=(8,3))
plt.scatter(hurricanes.Year, hurricanes.Error_72h,
            s=8, marker='o');
```

```python
plt.figure(figsize=(8,3))
plt.scatter(hurricanes.Year, hurricanes.Error_72h,
            s=8, marker='o');
smoothed_error_72h = sm.nonparametric.lowess(hurricanes.Error_72h, hurricanes.Year)
plt.plot(smoothed_error_72h[:, 0], smoothed_error_72h[:, 1], 'r');
```

```python
np.corrcoef(hurricanes.Error_72h, hurricanes.Year)
```
