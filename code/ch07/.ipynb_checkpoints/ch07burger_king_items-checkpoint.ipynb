```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import statistics as stat
```

```python
datafile = '../../dataset/ch07/burger_king_items.txt'
bk = pd.read_csv(datafile, sep='\t')
```

```python
bk.info
```

```python
len(bk)
```

```python
bk.columns
```

```python
fat = bk['Fat(g)']
protein = bk['Protein(g)']
```

```python
plt.scatter(protein, fat,
            s=8, marker='o');
smoothed = sm.nonparametric.lowess(fat, protein)
plt.plot(smoothed[:, 0], smoothed[:, 1], 'r');
```

```python
linear_fit = LinearRegression()
x = np.array(protein).reshape(-1, 1)
y = np.array(fat)
```

```python
linear_fit.fit(x, y)
```

```python
linear_fit.coef_, linear_fit.intercept_
```

```python
plt.scatter(x, y,
            s=8, marker='o');
plt.plot(x, linear_fit.predict(x), 'r');
```

```python
x_bar = x.mean()
y_bar = y.mean()
sd_x = x.std(ddof=1)
sd_y = y.std(ddof=1)
r = np.corrcoef(x.reshape(1, -1), y)[0, 1]
x_bar, y_bar, sd_x, sd_y, r
```

```python
b_1 = r * sd_y / sd_x
b_1
```

```python
b_0 = y_bar - b_1 * x_bar
b_0
```
