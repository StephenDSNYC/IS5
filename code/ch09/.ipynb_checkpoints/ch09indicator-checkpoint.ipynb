```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
coasters = '../../dataset/ch09/coasters_2015.txt'
coasters = pd.read_csv(coasters, sep='\t')
```

```python
coasters.head()
```

```python
coasters = coasters[(~coasters.Drop.isna()) & (~coasters.Duration.isna())]  # remove NA from the two columns
```

```python
x = coasters['Drop']
y = coasters['Duration']
```

```python
plt.scatter(x, y);
plt.xlabel('Drop')
plt.ylabel('Duration');
```

```python
coasters[coasters.Drop>300]  # see footnote 3 on page 288.  This data point is subsequently dropped
```

```python
coasters = coasters.drop(index=71)
```

```python
coasters[coasters.Name=='Xcelerator']  # see footnote 3 on page 288.  This data point is subsequently dropped
```

```python
coasters = coasters.drop(index=6)
```

```python
x = coasters['Drop']
y = coasters['Duration']

plt.scatter(x, y);
plt.xlabel('Drop')
plt.ylabel('Duration');
```

```python
linfit = LinearRegression()
```

```python
linfit.fit(x.values.reshape(-1, 1), y);
```

```python
plt.scatter(x, y, marker='o', color='b')
plt.plot(x, linfit.predict(x.values.reshape(-1, 1)), color='r')
plt.xlabel('Drop')
plt.ylabel('Duration');
```

```python
plt.scatter(y, x - linfit.predict(x.values.reshape(-1, 1)), marker='o', color='b')
plt.xlabel('Drop')
plt.ylabel('residues');
```

```python
coasters.Inversions.value_counts()
```

```python
x1 = coasters[coasters.Inversions==1]['Drop']
y1 = coasters[coasters.Inversions==1]['Duration']
x2 = coasters[coasters.Inversions==0]['Drop']
y2 = coasters[coasters.Inversions==0]['Duration']
```

```python
plt.scatter(x1, y1, color='r', marker='x', s=12)
plt.scatter(x2, y2, color='b', marker='o', s=16);
```

```python
plt.scatter(x1, y1, color='r', marker='x', s=12)
plt.scatter(x2, y2, color='b', marker='o', s=16)
linfit.fit(x1.values.reshape(-1, 1), y1)
plt.plot(x, linfit.predict(x.values.reshape(-1, 1)), color='r')
linfit.fit(x2.values.reshape(-1, 1), y2)
plt.plot(x, linfit.predict(x.values.reshape(-1, 1)), color='b');
```

```python
X = coasters[['Drop', 'Inversions']]
y = coasters['Duration']
```

```python
linfit.fit(X, y)
print(linfit.coef_)
plt.scatter(linfit.predict(X[X.Inversions==1]), y[X.Inversions==1]-linfit.predict(X[X.Inversions==1]), 
            marker='x', color='r', s=12)
plt.scatter(linfit.predict(X[X.Inversions==0]), y[X.Inversions==0]-linfit.predict(X[X.Inversions==0]), 
            marker='o', color='b', s=16);
```
