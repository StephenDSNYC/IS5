```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
datafile = '../../dataset/ch08/penguins.txt'
```

```python
penguins = pd.read_csv(datafile, sep='\t')
```

```python
penguins.info
```

```python
penguins.head()
```

```python
control = penguins['Duration(min)']
response = penguins['Dive Heart Rate']
```

```python
plt.figure(figsize=(8,3));
plt.scatter(control, response, color='maroon', s=20);
smoothed = sm.nonparametric.lowess(response, control)
plt.plot(smoothed[:, 0], smoothed[:, 1])
plt.xlabel('Duration (min)');
plt.ylabel('Dive Heart Rate (bpm)');
plt.xticks([4, 8, 12, 16]);
plt.yticks(list(range(30, 121, 30)));
plt.title('Dive Heart Rate ~ Duration');
```

```python
linfit = LinearRegression()
linfit.fit(np.array(control).reshape(-1,1), 
           np.array(response));
```

```python
plt.figure(figsize=(8,3));
plt.scatter(control, response, color='maroon', s=20);
plt.plot(control, linfit.predict(np.array(control).reshape(-1,1)), color='k', lw=2)
plt.xlabel('Duration (min)');
plt.ylabel('Dive Heart Rate (bpm)');
plt.xticks([4, 8, 12, 16]);
plt.yticks(list(range(30, 121, 30)));
plt.title('Dive Heart Rate ~ Duration w/ Linear Regression');
```

```python
r = np.corrcoef(control, response)[0, 1]
print(f'R_squared: {r ** 2 * 100:.1f}%')
```

```python
plt.figure(figsize=(8,3));
residues = response-linfit.predict(np.array(control).reshape(-1,1))
plt.scatter(control, residues, color='maroon', s=20);
smoothed = sm.nonparametric.lowess(residues, control)
plt.plot(smoothed[:, 0], smoothed[:, 1])
plt.xlabel('Duration (min)');
plt.ylabel('residues');
plt.xticks([4, 8, 12, 16]);
plt.title('Dive Heart Rate ~ Duration (Residues)');
```

**The parabola pattern shown by the LOWESS curve in the above plot indicates residual pattern not considered by the linear model.**

**Fitness of a linear model is thus questionable.**

```python
response2 = 60_000 / response
```

```python
plt.figure(figsize=(8,3));
plt.scatter(control, response2, color='maroon', s=20);
smoothed = sm.nonparametric.lowess(response2, control)
plt.plot(smoothed[:, 0], smoothed[:, 1])
plt.xlabel('Duration (min)');
plt.ylabel('Heart Beat Time (ms)');
plt.xticks([4, 8, 12, 16]);
plt.yticks([500, 1000, 1500, 2000, 2500, 3000]);
plt.title('1/Dive Heart Rate ~ Duration');
```

```python
linfit.fit(np.array(control).reshape(-1,1), 
           np.array(response2));
```

```python
plt.figure(figsize=(8,3));
plt.scatter(control, response2, color='maroon', s=20);
plt.plot(control, linfit.predict(np.array(control).reshape(-1,1)), color='k', lw=2)
plt.xlabel('Duration (min)');
plt.ylabel('Heart Beat Time (ms)');
plt.xticks([4, 8, 12, 16]);
plt.yticks([500, 1000, 1500, 2000, 2500, 3000]);
plt.title('1/Dive Heart Rate ~ Duration (Linear Regression)');
```

```python
r2 = np.corrcoef(control, response2)[0, 1]
print(f'R_squared: {r2 ** 2 * 100:.1f}%')
```

```python
plt.figure(figsize=(8,3));
residues = response2-linfit.predict(np.array(control).reshape(-1,1))
plt.scatter(control, residues, color='maroon', s=20);
smoothed = sm.nonparametric.lowess(residues, control)
plt.plot(smoothed[:, 0], smoothed[:, 1])
plt.xlabel('Duration (min)');
plt.ylabel('Heart Beat Time (ms)');
plt.xticks([4, 8, 12, 16]);
plt.title('1/Dive Heart Rate ~ Duration (Residues)');
```

The LOWESS curve does not show any clear pattern.  Our linear fitting is considered more appropriate.

**However, note that the residues are of wider spread in the middle than the two sides, possibly suggesting heteroscedacity.**

```python
response3 = 1/np.sqrt(response)
```

```python
plt.figure(figsize=(8,3));
plt.scatter(control, response3, color='maroon', s=20);
smoothed = sm.nonparametric.lowess(response3, control)
plt.plot(smoothed[:, 0], smoothed[:, 1])
plt.xlabel('Duration (min)');
plt.ylabel('1/sqrt(Dive Heart Rate)');
plt.xticks([4, 8, 12, 16]);
plt.title('1/sqrt(Dive Heart Rate) ~ Duration');
```

```python
linfit.fit(np.array(control).reshape(-1,1), 
           np.array(response3));
plt.figure(figsize=(8,3));
plt.scatter(control, response3, color='maroon', s=20);
plt.plot(control, linfit.predict(np.array(control).reshape(-1,1)), color='k', lw=2)
plt.xlabel('Duration (min)');
plt.ylabel('1/sqrt(Dive Heart Rate)');
plt.xticks([4, 8, 12, 16]);
plt.title('1/sqrt(Dive Heart Rate) ~ Duration (Linear Regression)');
```

```python
r3 = np.corrcoef(control, response3)[0, 1]
print(f'R_squared: {r3 ** 2 * 100:.1f}%')
```

```python
plt.figure(figsize=(8,3));
residues = response3-linfit.predict(np.array(control).reshape(-1,1))
plt.scatter(control, residues, color='maroon', s=20);
smoothed = sm.nonparametric.lowess(residues, control)
plt.plot(smoothed[:, 0], smoothed[:, 1])
plt.xlabel('Duration (min)');
plt.ylabel('1/sqrt(Dive Heart Rate)');
plt.xticks([4, 8, 12, 16]);
plt.title('1/sqrt(Dive Heart Rate) ~ Duration (Residues)');
```
