```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statistics as stat
```

```python
# 10 x sqrt
import math
def sqrtx10(x: float):  # a concave transformation
    return math.sqrt(x) * 10
sqrtx10 = np.vectorize(sqrtx10)
```

```python
x = np.linspace(0, 100, 1001)
y = sqrtx10(x)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y, 'b-', x, x, 'r-.');
ax.grid()
ax.set_aspect('equal')
ax.legend(['sqrtx10', 'id']);
```

```python
x = np.linspace(1, 100, 1001)
y = np.log(x)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y);
ax.grid()
ax.legend(['ln(x)']);
```

```python
x = np.linspace(0.1, 10, 401)
y = 1 / x
fig, ax = plt.subplots(1, 1)
ax.plot(x, y);
ax.grid()
ax.legend(['1/x']);
```
