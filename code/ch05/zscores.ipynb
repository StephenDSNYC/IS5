```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.normal(170, 10, 500)
```

```python
min(y), max(y)
```

```python
plt.hist(y, bins=20);
```

```python
numer = y - y.mean()
```

```python
plt.hist(numer, bins=20);
```

```python
z = numer/y.std(ddof=1)
```

```python
plt.hist(z, bins=20);
```
