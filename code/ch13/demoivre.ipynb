```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import sqrt
```

```python
def sim_bernoulli(p):
    """
    SIMulate a BERNOULLI trial with success probability p
    """
    if random.random() < p:
        return 1  # sucesss
    else:
        return 0  # failure
```

```python
def sim_bernoulli_n(p, n):
    """
    SIMulate n BERNOULLI trials with success probability p
    """
    return [sim_bernoulli(p) for i in range(n)]
```

```python
#set up parameters
p = 0.4
q = 1-p
num_sample = 20_000  # number of samples
sample_size = 2_000  # number of trials in each sample
sd = sqrt(p * q / sample_size)
proportions = np.zeros(num_sample)
#take samples and pool proportions
for i in range(num_sample):
    sample = sim_bernoulli_n(p, sample_size)
    proportions[i] = (sum(sample)/sample_size - p) / sd
#plot historam to approxmate density
plt.hist(proportions, bins=27, density=True);
#overlay Gaussian density
x_axis = np.linspace(-4, 4, 1_000)
plt.plot(x_axis, norm.pdf(x_axis));
```
