```python
import pandas as pd
```

```python
file_name = "../../dataset/ch02/credit_card_charges.txt"
```

```python
credit_card_charges= pd.read_csv(file_name, sep='\t')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.hist(credit_card_charges, bins=20, edgecolor='black');
```

```python
plt.hist(credit_card_charges, bins=np.linspace(0, 1000, 20), edgecolor='black');
```

```python
import seaborn as sns
```

```python
sns.kdeplot(credit_card_charges, fill=True, color='blue', bw_adjust=0.6);
```

```python
import statistics as stat
type(credit_card_charges)
```

```python
avg_chrgs = stat.mean(credit_card_charges["Charges($)"])
print(f"Average Credit Card Charges: {avg_chrgs:.2f}")
```

```python
# avg_chrgs = stat.mean(credit_card_charges.Charges($))  # this is a SyntaxError
# print(f"Average Credit Card Charges: {avg_chrgs:.2f}")
```

```python
med_chrgs = stat.median(credit_card_charges["Charges($)"])
print(f"Median Credit Card Charges: {med_chrgs:.2f}")
```

```python
charges_range = max(credit_card_charges["Charges($)"]) - min(credit_card_charges["Charges($)"])
print(f"range of Credit Card Charges: {charges_range}")
```

```python
import numpy as np
q75, q25 = np.percentile(credit_card_charges, [75, 25])
print(f"interquartile range of Credit Card Charges: {q75 - q25}")
```

```python
sd_credit_card_charges = stat.stdev(credit_card_charges["Charges($)"])
print(f"Standard Deviation of Credit Card Charges: {sd_credit_card_charges:.2f}")
```
