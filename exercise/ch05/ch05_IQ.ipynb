```python
# Problem 10(a)
# Draw the model for these IQ scores. Clearly label it, 
# showing what the 68-95-99.7 Rule predicts
# y_bar = 100
# s = 15

# below is a standard normal, not the one as required in this problem
mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

** Problem 10(c)**


About what precent of people should have IQ scores above 115?

Answer: According to the 68-95-99.7 Rule, it is about 1.5 precent of people should have IQ scores above 115.


```python
## Problem 10(c)
y_bar = 100
s = 15
y = 115
z = (y - y_bar)/s
p = stats.norm.cdf(z)  # scored under 115
print(f"It is about {(1-p)*100:.0f} precent of people"
       " should have IQ scores above 115.")
```

** Problem 10(d)


About what precent of people should have IQ scores between 70 and 85?

Answer: According to the 68-95-99.7 Rule, It is about 1.5 precent of people should have IQ scores between 70 and 85.

```python
y_bar = 100
s= 15

y1 = 70
z1 = (y1 - y_bar)/s

y2 = 85 
z2 = (y2 - y_bar)/s

p1 = stats.norm.cdf(z1)
p2 = stats.norm.cdf(z2)

print(f"About {(p2-p1)*100:.1f} precent of people"
       " should have IQ scores between 70 and 85.")
```

** 10(f) **
Suppose I want to score in the first 10% of all test takers, what is the minimum score I should achieve?

** Problem 38**

Question: Exercise 10 proposes modeling IQ scores with N(100, 15).  What IQ would you consider to be unusually high? Explain.


```python
## Problem 38##
y_bar = 100
s = 15

z = stats.norm.ppf(0.9985)  # corresponds to the 3 sigma rule

y = y_bar + z*s

print(f"I would consider {y:.1f} is unusually high.")
```

### Note:

In statistics, we typically consider 5% of the population as extreme cases.  In some occasions, such percentage may be further reduced down to 2%, 1% or .5%.

In above analysis, the student condisers .5% (equally divived to both tails) as extreme.

Alternatively, if 5% is used as a threshold, she may conclude that 2 sigma above the average (i.e., 130 in IQ score or higher) can already be considered as a genius.

### Problem 50


In the Normal model N(100, 15) from Exercise 10, what cutoff value bounds


(a) the hightest 5% of all IQs?

(b) the lowest 30% of the IQs?

(c) the middle 80% of the IQs?


```python
## Problem 50(a)##
y_bar = 100
s = 15

z = stats.norm.ppf(0.95)
y = y_bar + z * s

print(f" The cutoff value is {y:.0f} of the hightest 5% of all IQs.")
```

```python
# Problem 50 (b)

y_bar = 100
s = 15

z = stats.norm.ppf(0.30)
y = y_bar + z * s

print(f" The cutoff value is {y:.0f} of the lowest 30% of the IQs.")
```

```python
# Problem 50 (c)

y_bar = 100
s = 15

p1 = 0.1
p2 = 0.9

z1 = stats.norm.ppf(p1)
z2 = stats.norm.ppf(p2)

y1 = y_bar + z1 *s
y2 = y_bar + z2 *s

print(f"The middle 80% of the IQs fall between {y1:.0f} and {y2:.0f}.")
```

### Problem 52


Consider the IQ model N(100, 15) one last time.

(a) What IQ represents the 15th percentile?

(b) What IQ represents the 98th percentile?

(c) What's the IQR of the IQs?

```python
# Problem 52 (a)

y_bar = 100
s = 15

z = stats.norm.ppf(0.15)
y = y_bar + z*s

print(f" {y:.0f} represents the 15th percentile.")
```

```python
# Problem 52 (b)

y_bar = 100
s = 15

z = stats.norm.ppf(0.98)
y = y_bar + z*s

print(f" {y:.0f} represents the 98th percentile.")
```

```python
# Problem 52 (c)

y_bar = 100
s = 15

z1 = stats.norm.ppf(0.25)
z2 = stats.norm.ppf(0.75)

y1 = y_bar + z1*s
y2 = y_bar + z2*s

print(f" {(y2-y1):.0f} is the IQR of the IQs.")
```
